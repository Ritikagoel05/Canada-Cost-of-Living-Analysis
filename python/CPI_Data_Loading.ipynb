```python
import pandas as pd
```

```python
df_raw = pd.read_csv("../data/cpi_raw.csv", skiprows=8)
```

```python
df_raw.head()
```

```python
# Remove the first two messy rows (index 0 and 1)
df_cleaned = df_raw.drop([0, 1])
```

```python
df_cleaned.head()
```

```python
df_cleaned.columns = df_raw.iloc[0]
```

```python
df_cleaned = df_cleaned.drop(df_cleaned.index[0])
```

```python
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
df_cleaned.head()
```

```python
df_melted = df_cleaned.melt(id_vars=['Products and product groups 3 4'], 
                            var_name='Date', 
                            value_name='CPI_Value')
```

```python
df_melted = df_melted.rename(columns={
    'Products and product groups 3 4': 'Product'
})
```

```python
df_melted.head(10)
```

```python
# Keep only rows where CPI_Value can be converted to a number
df_melted_clean = df_melted[pd.to_numeric(df_melted['CPI_Value'], errors='coerce').notnull()]
```

```python
df_melted_clean['CPI_Value'] = df_melted_clean['CPI_Value'].replace(",", ".", regex=True)
df_melted_clean['CPI_Value'] = df_melted_clean['CPI_Value'].astype(float)
```

```python
df_melted_clean = df_melted_clean.reset_index(drop=True)
```

```python
df_melted_clean.head(10)
```

```python
df_melted_clean.to_csv("../data/cpi_cleaned.csv", index=False)
```
